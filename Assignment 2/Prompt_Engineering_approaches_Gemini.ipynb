Step 1: Installation and Configuration

```python
import google.generativeai as genai
from google.colab import userdata

Gemini_api_key = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=Gemini_api_key)

# Initialize the model (Gemini 2.5 Flash is ideal for general text/reasoning tasks)
model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
def get_response(prompt):
    """Helper function to generate and print the response."""
    response = model.generate_content(prompt)
    print(response.text)
    print("-" * 50)
```

Step 2: Zero-Shot vs. Few-Shot Prompting

```python
print("=== ZERO-SHOT PROMPTING ===")
zero_shot_prompt = """
Classify the severity of the following tomato plant observation into 'Low', 'Medium', or 'High'.

Observation: The lower leaves are exhibiting slight yellowing, but the overall stem structure remains rigid and there is no sign of wilting or fungal spots.
Severity:
"""
get_response(zero_shot_prompt)


print("=== FEW-SHOT PROMPTING ===")
few_shot_prompt = """
Classify the severity of the following tomato plant observation into 'Low', 'Medium', or 'High'.

Observation: Rapid wilting of the entire plant within 24 hours, accompanied by a white fungal growth near the base of the stem.
Severity: High

Observation: A few small, dark concentric spots have appeared on the older leaves, but the rest of the canopy is green.
Severity: Medium

Observation: The lower leaves are exhibiting slight yellowing, but the overall stem structure remains rigid and there is no sign of wilting or fungal spots.
Severity:
"""
get_response(few_shot_prompt)
```

Step 3: Chain of Thought (CoT) Prompting

```python
print("=== CHAIN OF THOUGHT (CoT) PROMPTING ===")
# We use the magic phrase "Think step-by-step" or explicitly map out the reasoning requirement.
cot_prompt = """
A trader sets up a predictive algorithm with $10,000.
On Day 1, the algorithm buys Stock A at $50 per share.
On Day 2, Stock A drops by 20%, so the algorithm buys another $2,000 worth of shares at the new price.
On Day 3, the stock price rises to $60 per share, and the algorithm sells all holdings.

Calculate the total net profit or loss. Think step-by-step and show your calculations before giving the final answer.
"""
get_response(cot_prompt)
```

Step 4: Tree of Thoughts (ToT) Prompting

```python
print("=== TREE OF THOUGHTS (ToT) PROMPTING ===")
tot_prompt = """
Imagine three different AI architects are proposing a tech stack for a new full-stack application that needs to handle real-time predictive analytics.

Step 1: Architect 1, Architect 2, and Architect 3 each propose a distinctly different frontend, backend, and deployment stack.
Step 2: A Senior Engineer evaluates the pros and cons of each architecture, specifically focusing on latency, scalability, and ease of deployment.
Step 3: The Senior Engineer selects the best overall architecture and explains why.

Present this process clearly.
"""
get_response(tot_prompt)
```

Step 5: The Interview Approach (Persona/Flipped Interaction)

```python
print("=== THE INTERVIEW APPROACH (INTERACTIVE CHAT) ===")

# Start a conversational session with the model
chat = model.start_chat(history=[])

# The initial instruction sets the persona and the rules for the AI
initial_prompt = """
Act as an expert Academic Advisor and Research Mentor. I want to write a high-impact book chapter on IoT integration.
Do not give me a topic right away. Instead, ask me 1 targeted question at a time about my existing knowledge, datasets I have access to, or my target audience.
Wait for my answer before asking the next question or suggesting a title. Ask your first question now.
"""

# Send the initial setup and get the AI's first question
response = chat.send_message(initial_prompt)
print(f"AI Mentor: {response.text}\n")

# Run an interactive loop for 3 turns (you can increase the range for a longer chat)
for i in range(3):
    user_input = input("You: ")

    # Allow a way to exit the loop early
    if user_input.lower() in ['exit', 'quit', 'stop']:
        print("Ending interview...")
        break

    # Send your answer and get the AI's follow-up
    response = chat.send_message(user_input)
    print(f"\nAI Mentor: {response.text}\n")
```
