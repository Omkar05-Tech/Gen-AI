Option 1: Implementation

```python
import torch
from diffusers import StableDiffusionPipeline

# Load model (uses Transformer + diffusion internally)
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

# Use GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = pipe.to(device)

# Prompt (You can change this)
prompt = "A futuristic smart city with flying cars, ultra realistic, 4K, cinematic lighting"

# Generate image
image = pipe(prompt).images[0]

# Save output
image.save("generated_image.png")

print("✅ Image generated and saved as generated_image.png")
```

Option 2: Advanced (Control + Better Results)

```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")

prompt = "A cyberpunk city at night, neon lights, rain, ultra detailed"

image = pipe(
    prompt,
    guidance_scale=7.5,   # controls prompt importance
    num_inference_steps=50
).images[0]

image.save("advanced_output.png")
```
